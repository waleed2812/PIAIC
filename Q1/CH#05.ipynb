```python
lines = int(input("Enter number of lines: "))

for i in range(0,lines,1):
    for j in range(lines,1,-1):
        print(" ",end=" ")
    
    for j in range(i+1,1,-1):
        print(j,end=" ")
        
    print("1",end=" ")
    
    for j in range(2,i+2,1):
        print(j,end=" ")
    lines -= 1
    print()
```
